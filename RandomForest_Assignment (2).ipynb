Split Dataset

```python
x=df.iloc[ : , :-1]
y=df.iloc[ : ,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=45)
```

```python
xtrain
```

```python
ytrain
```

```python
xtest
```

```python
ytest
```

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
xtrain_scaled=std.fit_transform(xtrain)
xtrain_scaled
```

```python
xtest_scaled=std.fit_transform(xtest)
xtest_scaled
```

# Random Forest Model Building

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
Rmodel=RandomForestClassifier(n_estimators=100, random_state=10)
Rmodel.fit(xtrain_scaled, ytrain)
```

```python
y_pred=Rmodel.predict(xtest_scaled)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,precision_score,f1_score,recall_score
```

```python
print('accuracy=',accuracy_score(ytest,y_pred)*100)
```

```python
print('Precision=',precision_score(ytest,y_pred,average='weighted')*100)
print('F1=',f1_score(ytest,y_pred,average='weighted')*100)
print('Recall=',recall_score(ytest,y_pred,average='weighted')*100)
```

# Bagging and Boosting

```python
from sklearn.ensemble import BaggingClassifier,AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
#Bagging
cart=DecisionTreeClassifier()
bagg=BaggingClassifier(estimator=cart,n_estimators=100,random_state=0)
```

```python
bagg.fit(xtrain_scaled,ytrain)
```

```python
ypred_Ba=bagg.predict(xtest_scaled)
ypred_Ba
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest,ypred_Ba))
```

```python
#Boosting
boost=AdaBoostClassifier(estimator=cart,n_estimators=100,random_state=2)
boost.fit(xtrain_scaled, ytrain)
```

```python
ypred_bo=boost.predict(xtest_scaled)
```

```python
print(classification_report(ytest,ypred_bo))
```
